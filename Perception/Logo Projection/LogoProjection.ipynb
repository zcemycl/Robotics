#### Homography Estimation
To project one image patch onto another, for each point inside the goal in the video frame, to find the corresponding point from the logo image to copy over. In the other words, the homography between the two image patches is required to compute. The homography is a $3\times 3$matrix that satisfies the following:
$$\mathrm{x}_{logo}\sim H \mathrm{x}_{video}$$
Or, equivalently:
$$\lambda\mathrm{x}_{logo}\sim H \mathrm{x}_{video}$$
where $\mathrm{x}_{logo}$ and $\mathrm{x}_{video}$ are homogeneous image coordinates from each patch and $\lambda$ is some scaling constant

A homography $H$ maps a set of points $\mathrm{x}=(x,y,1)^\top$ to another set of points $\mathrm{x}^\prime = (x',y',1)^\top$ up to a scalar.

\begin{align}
\mathrm{x}^\prime &\sim H\mathrm{x}\\
\lambda
\begin{pmatrix}
x'\\ y'\\ 1
\end{pmatrix} &= \begin{pmatrix}
h_{11} & h_{12} & h_{13}\\
h_{21} & h_{22} & h_{23}\\
h_{31} & h_{32} & h_{33}\\
\end{pmatrix}
\begin{pmatrix}
x \\ y \\ 1
\end{pmatrix}\\
\lambda x' &= h_{11}x+h_{12}y+h_{13}\\
\lambda y' &= h_{21}x+h_{22}y+h_{23}\\
\lambda &= h_{31}x+h_{32}y+h_{33}
\end{align}


Therefore, $x'$ and $y'$ are given by,

\begin{align}
x' &= \frac{h_{11}x+h_{12}y+h_{13}}{h_{31}x+h_{32}y+h_{33}} = \frac{h_{11}x_1+h_{12}x_2+h_{13}}{h_{31}x_1+h_{32}x_2+h_{33}} = x_1'\\
y' &= \frac{h_{21}x+h_{22}y+h_{23}}{h_{31}x+h_{32}y+h_{33}} = \frac{h_{21}x_1+h_{22}x_2+h_{23}}{h_{31}x_1+h_{32}x_2+h_{33}} = x_2'
\end{align}

Rearranging the terms above, 

\begin{align}
-h_{11}x_1-h_{12}x_2-h_{13}+h_{31}x_1x_1'+h_{32}x_2x_1'+h_{33}x_3x_1'&= 0 \\
-h_{21}x_1-h_{22}x_2-h_{23}+h_{31}x_1x_2'+h_{32}x_2x_2'+h_{33}x_3x_2'&= 0 \\
\begin{pmatrix} a_x\\a_y \end{pmatrix} h &= 0 
\end{align}

where 

\begin{align}
a_x &= (-x_1,-x_2,-1,0,0,0,x_1x_1',x_2x_1',x_1') \\
a_y &= (0,0,0,-x_1,-x_2,-x_3,x_1x_2',x_2x_2',x_2') \\
h &= (h_{11},h_{12},h_{13},h_{21},h_{22},h_{23},h_{31},h_{32},h_{33})^\top
\end{align}

Matrix $H$ has a degree of freedom, as each point gives 2 sets of equations, so 4 points are required to solve for $h$ uniquely.By stacking,
$$\mathrm{A} = \begin{pmatrix}
a_{x1}\\a_{y1}\\ \vdots \\ a_{xn}\\a_{yn}\\
\end{pmatrix}$$
Solving the following equations, $$\mathrm{A}h=0$$
The vector $h$ is the last column of $V$ where $\mathrm{A}=U\Sigma V^\top$.

```python
def calculateInteriorPts(img_size,corners):
    x,y = np.meshgrid(np.linspace(1,img_size[1],img_size[1]),
                     np.linspace(1,img_size[0],img_size[0]))
    X=x.reshape(-1,1); Y=y.reshape(-1,1)
    
    p = path.Path(corners)
    allpts = np.array([X,Y]).squeeze().T
    interior_bol=p.contains_points(allpts)
    return allpts[interior_bol,:]
```

```python
tmp = cv2.imread(readdir.format(test_images[0]))
tmpvidpts = video_pts[:,:,0]
interior_pts = calculateInteriorPts(tmp.shape,tmpvidpts)
plt.imshow(tmp)
plt.plot(interior_pts[:,0],interior_pts[:,1],'r.')
```

```python
def vectorize(r,rp):
    x1=r[0]; x2=r[1]; x3=1
    x1p=rp[0]; x2p=rp[1]
    ax=[-x1,-x2,-1,0,0,0,x1*x1p,x2*x1p,x1p]
    ay=[0,0,0,-x1,-x2,-x3,x1*x2p,x2*x2p,x2p]
    return ax,ay
```

```python
def est_homography(video_pts,logo_pts):
    A = []
    for i in range(len(logo_pts)):
        r = video_pts[i,:]; rp = logo_pts[i,:]
        ax,ay = vectorize(r,rp)
        A.append(ax)
        A.append(ay)
    A = np.array(A)
    u, s, vh = np.linalg.svd(A, full_matrices=True)
    H = vh[-1,:].reshape((3,3))
    return H
```

```python
H = est_homography(tmpvidpts,logo_pts)
H
```

```python
def warp_pts(video_pts,logo_pts,sample_pts):
    H = est_homography(video_pts,logo_pts)
    warped_pts = []
    for i in range(len(sample_pts)):
        tmp = np.concatenate((sample_pts[i,:],np.array([1])),axis=0)
        result = np.matmul(H,tmp)
        warped_pts.append(result[0:2]/result[2])
    return np.array(warped_pts)
```

```python
warped_logo_pts=warp_pts(tmpvidpts,logo_pts,interior_pts)
```

```python
def inverse_warping(imgf,imgi,ptsf,ptsi):
    ptsf = np.floor(ptsf).astype(int)
    ptsi = np.floor(ptsi).astype(int)
    projected_img = imgf
    for i in range(3):
        sif = imgf[:,:,i]
        sii = imgi[:,:,i]
        sif[ptsf[:,1],ptsf[:,0]]=sii[ptsi[:,1],ptsi[:,0]]*.5+sif[ptsf[:,1],ptsf[:,0]]*.5
        projected_img[:,:,i]=sif
    return projected_img
```

```python
result = inverse_warping(tmp,image,interior_pts,warped_logo_pts)
```
